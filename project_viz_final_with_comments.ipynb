```python
from bokeh.models import HoverTool
from bokeh.sampledata.periodic_table import elements
hv.extension('bokeh')
def create_holoplot(label, xlim= (0,4000)):
    selected_trace = merged_df[label]
    # for i in points:
    #     print(i)                                                                
    plot = hv.Scatter((selected_trace,merged_df['Premature Deaths by Population'], selected_trace, merged_df['State']), label=label, kdims=[label,'Premature Deaths by Population'], vdims=['State'])
    if label == 'Adult Obesity':
        color = 'red'
    elif label == 'Driving Alone':
        color= 'green'
    else:
        color= 'black'
    plot.opts(  title=f'For West Virginia Counties {label} vs Premature Deaths in state', height= 500,width=800,  xrotation=90,color=color,tools=['hover'])
    return plot

# Create a Select widget
label_selector = pn.widgets.Select(name='FACTOR', options=['Adult Obesity', 'Driving Alone', 'Excessive Adult Smoking'])

# Use pn.interact to update the HoloViews plot based on the selected label
holoplot_pane = pn.panel(pn.interact(create_holoplot, label=label_selector))

holoplot_pane.servable()
```
